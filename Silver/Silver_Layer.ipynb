Upload Data

```python
import pandas as pd

github_url = "https://raw.githubusercontent.com/DuaA-A/Enterprise-E-Commerce-Predictive-Analytics/main/DailySales%20(1).xlsx"

df = pd.read_excel(github_url)

print("Shape:", df.shape)
df.head()
```

Data Cleaning

```python

print(df.info())
display(df.describe())
print(df.isnull().sum())
```

Remove Duplicates

```python
print("Duplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("Shape After Removing Duplicates:")
print(df.shape)
```

Inconsistent

```python
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

for col in text_cols:
    df[col] = df[col].str.title()

print("Inconsistent Values Fixed")
```

Null

```python

null_values = df.isnull().sum()

print(null_values[null_values > 0])
```

Missing Data

```python


numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

print("Missing Values Handled")
```

Rename Columns

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

print(df.columns.tolist())
```

Standardization

```python
for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors='coerce')

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

print("Standardization Completed")
```

Save Silver Layer

```python
output_file = "DailySales_Silver.xlsx"

df.to_excel(output_file, index=False)

print("Silver Layer Saved Successfully")
```
