Upload Data

```python
import pandas as pd
import numpy as np

github_url = "https://raw.githubusercontent.com/DuaA-A/Enterprise-E-Commerce-Predictive-Analytics/main/Bronze/raw_data.csv"

df = pd.read_csv(github_url)

print("Shape:", df.shape)
df.head()
```

Data Cleaning

```python

print(df.info())
print(df.info())
print(df.isnull().sum())

empty_ratio = df.isnull().mean(axis=1)
junk_rows = empty_ratio > 0.9
print("Junk/near-empty rows removed:", junk_rows.sum())
df = df[~junk_rows].reset_index(drop=True)

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Date"]).reset_index(drop=True)
print("Shape after removing junk rows:", df.shape)
display(df.describe())
print(df.isnull().sum())
```

Rename Columns

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace("%", "pct")
)
print(df.columns.tolist())
```

Duplicate

```python
print("Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print("Shape After Removing Duplicates:", df.shape)

fully_empty_cols = [c for c in df.columns if df[c].isnull().all()]
print("Fully empty columns dropped:", fully_empty_cols)
df = df.drop(columns=fully_empty_cols)
```

Inconsistent

```python
id_cols = [c for c in ["sku", "fnsku", "asin", "parent_asin", "market_place"] if c in df.columns]
text_cols = df.select_dtypes(include="object").columns.tolist()
descriptive_cols = [c for c in text_cols if c not in id_cols]

for col in id_cols:
    df[col] = df[col].astype("string").str.strip()

for col in descriptive_cols:
    df[col] = df[col].astype("string").str.strip().str.title()

print("Inconsistent Values Fixed")
```

Null

```python
null_values = df.isnull().sum()
print(null_values[null_values > 0])
```

Missing Data

```python
count_like_cols = [c for c in [
    "orders", "units", "refunded", "promo_units", "organic_units",
    "sessions", "page_views", "impressions", "clicks", "ppc_orders"
] if c in df.columns]

for col in count_like_cols:
    df[col] = df[col].fillna(0)

other_numeric_cols = [c for c in df.select_dtypes(include="number").columns if c not in count_like_cols]
for col in other_numeric_cols:
    df[col] = df[col].fillna(df[col].median())

categorical_cols = df.select_dtypes(include=["object", "string"]).columns
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

print("Missing Values Handled")
print("Remaining nulls:", df.isnull().sum().sum())
```

Standardization

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")

for col in df.select_dtypes(include=["object", "string"]).columns:
    df[col] = df[col].astype("string").str.strip()

for col in ["refund_pct", "unit_session_pct"]:
    if col in df.columns:
        out_of_range = ((df[col] < 0) | (df[col] > 1)).sum()
        print(f"{col}: {out_of_range} out-of-range values clipped to [0,1]")
        df[col] = df[col].clip(0, 1)

for col in ["net_margin", "net_roi"]:
    if col in df.columns:
        extreme = (df[col].abs() > 5).sum()
        print(f"{col}: {extreme} extreme/invalid values fixed")
        df.loc[df[col].abs() > 5, col] = np.nan
        df[col] = df[col].fillna(df[col].median())

print("Standardization Completed")
print("Final shape:", df.shape)
print("Final nulls:", df.isnull().sum().sum())
print("Final duplicates:", df.duplicated().sum())
```

Save Silver Layer

```python
output_file = "DailySales_Silver_Final.xlsx"
df.to_excel(output_file, index=False)
print("Silver Layer Saved Successfully ->", output_file)
```
